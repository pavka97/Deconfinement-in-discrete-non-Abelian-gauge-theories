```python
N = 4
d = 2*N

U = np.zeros((d,d,2,2),dtype = complex)
```

```python
#defining matrix elements of U
r_gen = np.array([[np.exp(2*np.pi/N*1j), 0], [0, np.exp(-2*np.pi/N*1j)]])
s_gen = np.array([[0, 1], [1, 0]])

for i in range(d):
    U[i][i] = np.linalg.matrix_power(s_gen, int(np.floor(i/N)))@np.linalg.matrix_power(r_gen, i)
    print(U[i][i]
          )
```

```python
#from group basis to representation basis (group Fourier transform)

QFTD3 = np.array([[1/np.sqrt(6), 1/np.sqrt(6), 1/np.sqrt(6), 1/np.sqrt(6), 1/np.sqrt(6), 1/np.sqrt(6)], #trivial
            [1/np.sqrt(6), 1/np.sqrt(6), 1/np.sqrt(6), -1/np.sqrt(6), -1/np.sqrt(6), -1/np.sqrt(6)], #parity
                        
            [1/np.sqrt(3), 1/np.sqrt(3)*np.exp(2*np.pi*1j/3), 1/np.sqrt(3)*np.exp(-2*np.pi*1j/3), 0, 0, 0],
            [0, 0, 0, 1/np.sqrt(3), 1/np.sqrt(3)*np.exp(-2*np.pi*1j/3), 1/np.sqrt(3)*np.exp(2*np.pi*1j/3)],
            [0, 0, 0, 1/np.sqrt(3), 1/np.sqrt(3)*np.exp(2*np.pi*1j/3), 1/np.sqrt(3)*np.exp(-2*np.pi*1j/3)],
            [1/np.sqrt(3), 1/np.sqrt(3)*np.exp(-2*np.pi*1j/3), 1/np.sqrt(3)*np.exp(2*np.pi*1j/3), 0, 0, 0]
                  ]) 

QFTD4 = np.array([
            [1/np.sqrt(8), 1/np.sqrt(8), 1/np.sqrt(8), 1/np.sqrt(8), 1/np.sqrt(8), 1/np.sqrt(8), 1/np.sqrt(8), 1/np.sqrt(8)],
            [1/np.sqrt(8), 1/np.sqrt(8), 1/np.sqrt(8), 1/np.sqrt(8), -1/np.sqrt(8), -1/np.sqrt(8), -1/np.sqrt(8), -1/np.sqrt(8)],
            [1/np.sqrt(8), -1/np.sqrt(8), 1/np.sqrt(8), -1/np.sqrt(8), 1/np.sqrt(8), -1/np.sqrt(8), 1/np.sqrt(8), -1/np.sqrt(8)],
            [1/np.sqrt(8), -1/np.sqrt(8), 1/np.sqrt(8), -1/np.sqrt(8), -1/np.sqrt(8), 1/np.sqrt(8), -1/np.sqrt(8), 1/np.sqrt(8)],
            [0.5 + 0j, 0 + 0.5j, -0.5 + 0j, 0 - 0.5j, 0 + 0j, 0 + 0j, 0 + 0j, 0 + 0j],
            [0 + 0j, 0 + 0j, 0 + 0j, 0 + 0j, 0.5 + 0j, 0 - 0.5j, -0.5 + 0j, 0 + 0.5j],
            [0 + 0j, 0 + 0j, 0 + 0j, 0 + 0j, 0.5 + 0j, 0 + 0.5j, -0.5 + 0j, 0 - 0.5j],
            [0.5 + 0j, 0 - 0.5j, -0.5 + 0j, 0 + 0.5j, 0 + 0j, 0 + 0j, 0 + 0j, 0 + 0j]
            ], dtype=complex)

if N == 3:
    QFT = QFTD3
elif N == 4:
    QFT = QFTD4

#nomenclature:
#reps: trivial, parity, fundamental rep.
```

```python
def corner_operators_rishons(i,j,pos):
    
    corner_op = np.zeros((rish_dim**2,rish_dim**2),dtype = complex)
    
    if pos%4 == 0:

        for k in range(2):
            
            corner_op += np.kron(rishons[i][k].T.conj(),rishons[j][k])
            
    elif pos%4 == 1:

        for k in range(2):
        
            corner_op += np.kron(rishons[i][(k+1)%2],rishons[j][k].T.conj())        
            #corner_op += np.kron(rishons[i][k],rishons[j][k].T.conj())        

    elif pos%4 == 2:

        for k in range(2):
        
            corner_op += np.kron(rishons[i][(k+1)%2].T.conj(),rishons[j][k])
            #corner_op += np.kron(rishons[i][k].T.conj(),rishons[j][k])
    
    elif pos%4 == 3:

        for k in range(2):
        
            corner_op += np.kron(rishons[i][k],rishons[j][k].T.conj())
            
    return corner_op
```

```python
def corner_operators(i,j,pos):
    
    corner_op = np.zeros((rish_dim**3,rish_dim**3),dtype = complex)
    
    
    if pos%4 == 0:

        for k in range(2): #Left Left
            corner_op += np.kron(np.eye(rish_dim),np.kron(rishons[i][k].T.conj(),rishons[j][k]))
            
    elif pos%4 == 1:

        for k in range(2):  #Right Left
        
            corner_op += np.kron(np.eye(rish_dim),np.kron(rishons[i][(k+1)%2],rishons[j][k].T.conj()))        
            #corner_op += np.kron(np.eye(rish_dim),np.kron(rishons[i][k],rishons[j][k].T.conj()))        

    
    elif pos%4 == 2:

        for k in range(2):  #Right Left
        
            corner_op += np.kron(np.kron(rishons[i][(k+1)%2].T.conj(),rishons[j][k]),np.eye(rish_dim))
            #corner_op += np.kron(np.kron(rishons[i][k].T.conj(),rishons[j][k]),np.eye(rish_dim))

            
    
    elif pos%4 == 3:

        for k in range(2):  #Right Right
            corner_op += np.kron(np.kron(rishons[i][(k+1)%2],rishons[j][(k+1)%2].T.conj()),np.eye(rish_dim))
            
            
    return corner_op
```

```python
def corner_operators_edge(i,j,pos,charge):
    
    if pos%4 == 0:
        
        corner_op = np.zeros((rish_dim**2,rish_dim**2),dtype = complex)

        for k in range(2):
            corner_op += np.kron(rishons[i][k].T.conj(),rishons[j][k])
            
    elif pos%4 == 1:
        if charge%2 == 1:
            corner_op = np.zeros((2*rish_dim**2,2*rish_dim**2),dtype = complex)


            for k in range(2):

                corner_op += np.kron(np.kron(rishons[i][(k+1)%2],rishons[j][k].T.conj()),np.eye(2))
                #corner_op += np.kron(np.kron(rishons[i][k],rishons[j][k].T.conj()),np.eye(2))

        else:
            corner_op = np.zeros((rish_dim**2,rish_dim**2),dtype = complex)


            for k in range(2):

                corner_op += np.kron(rishons[i][(k+1)%2],rishons[j][k].T.conj())           
                #corner_op += np.kron(rishons[i][k],rishons[j][k].T.conj())           


    elif pos%4 == 2:
        corner_op = np.zeros((rish_dim**2,rish_dim**2),dtype = complex)

        for k in range(2):
            corner_op += np.kron(rishons[i][(k+1)%2].T.conj(),rishons[j][k])
            #corner_op += np.kron(rishons[i][k].T.conj(),rishons[j][k])

            
    
    elif pos%4 == 3:
        if charge%2 == 1:

            corner_op = np.zeros((2*rish_dim**2,2*rish_dim**2),dtype = complex)

            for k in range(2):
                corner_op += np.kron(np.kron(rishons[i][(k+1)%2],rishons[j][(k+1)%2].T.conj()),np.eye(2))
                
        else:
            corner_op = np.zeros((rish_dim**2,rish_dim**2),dtype = complex)

            for k in range(2):
                corner_op += np.kron(rishons[i][(k+1)%2],rishons[j][(k+1)%2].T.conj())
            
    return corner_op
```
